```python
import pandas as pd 
import numpy as np 
import warnings
warnings.filterwarnings("ignore")
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies[['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'crew']]
```

```python
# handling missing data 
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
import ast
```

```python
# using literal_eval to convert string to list
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
def convert_cast(obj):
    L = []
    counter = 0
    
    if isinstance(obj, str):
        obj = ast.literal_eval(obj)
    
    for i in obj:
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
            
    return L

movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# transformation

movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
# vectorization converts data (text, images, audio) into numerical vectors 
#for AI comprehension or transforms pixel-based images (JPG/PNG) into scalable vector graphics (SVG)

from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = [] 
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
print(list(cv.get_feature_names_out()))
```

```python
# to calculate distance of all vectors from each other 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# using enumerate to write the movie index and its distance from the given movie index in a tuple

def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]  # fetching the index of the movie title
    distances = similarity[movie_index]   # calculating the distance of the given movie index with all other movies
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:6]  #lambdax:x[1] so that sorting occrs based on distance

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
